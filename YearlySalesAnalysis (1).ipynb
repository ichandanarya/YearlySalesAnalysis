```python
import pandas as pd
import os
import numpy as np
```

```python
path = "./Sales_Data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data.csv", index=False)
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data.head()
```

```python
# Extract first two characters as month text
all_data['Month'] = all_data['Order Date'].astype(str).str[0:2]

# Convert to numeric, forcing bad values to NaN instead of errors
all_data['Month'] = pd.to_numeric(all_data['Month'], errors='coerce')

# Drop rows where month is NaN (invalid dates)
all_data = all_data.dropna(subset=['Month'])

# Convert month to integer
all_data['Month'] = all_data['Month'].astype(int)

all_data.to_csv("all_data1.csv", index=False)
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'].astype('int') * all_data['Price Each'].astype('float')
all_data
```

```python
all_data.groupby('Month')['Sales'].sum(numeric_only=True)
```

```python
import matplotlib.pyplot as plt

months = range(1,13)
print(months)

plt.bar(months,all_data.groupby(['Month']).sum()['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
all_data.head()
```

```python
# Example: Total sales per city
city_sales = all_data.groupby('City')['Sales'].sum()
city_sales

city_sales.to_csv("city_sales.csv", index = False)
```

```python
import matplotlib.pyplot as plt

# Group sales by city
results = all_data.groupby('City').sum(numeric_only=True)

# Extract city names
cities = results.index

# Plot
plt.bar(cities, results['Sales'])
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.xticks(cities, rotation='vertical', size=8)
plt.show()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'], errors='coerce')
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data.head()
```

```python
# Group by hour and count orders
hours = all_data.groupby('Hour').count()['Order ID']   # or any column

# x-axis = hours (index), y-axis = counts
plt.plot(hours.index, hours.values)

plt.xticks(hours.index)   # tick marks at each hour
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

```python
df2 = all_data[all_data['Order ID'].duplicated(keep=False)].copy()

df2['Grouped'] = df2.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df2[['Order ID', 'Grouped']].drop_duplicates()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

# Top 10 most common product pairs
for key, value in count.most_common(10):
    print(key, value)
```
